### [3] 연월별 거래량 구하기

Transaction 칼럼은 거래번호를 나타냅니다. <br>
예를 들어 Transaction 2는 Scandinavian 품목 2개를 구매했다는 것을 의미합니다. <br>
그러면 연월별 거래량을 구하고 날짜 순으로 정렬해봅시다. <br>

<b> Hint </b> <br>
list 대신에 set을 이용하면 distinct한 element만 얻을 수 있습니다.<br>
연월별로 distinct한 transaction 수가 거래량을 의미하겠죠?! <br>

```python
# result1 = bread.map(lambda x: x[0][0:7]).collect()
```

```python
result1 = bread.groupBy(lambda element:element[0][0:7]).collect()
result1 = sorted(result1, key=lambda tup: tup[0])
for elements in result1:
  print(elements[0], len(set(list(map(lambda x: x[2], list(elements[1]))))))
```

### [4] 시간대별 총 판매량 구하기

시간대별 총 판매량을 구하고 판매량 순으로 정렬해봅시다. <br>
(거래량과 판매량은 다릅니다! 한 번의 거래에 여러 개를 판매할 수 있기 때문에...)

```python
result2 = bread.groupBy(lambda element:element[1][:2]).collect()
result2 = sorted(result2, key=lambda tup: len(tup[1]), reverse=True)
for elements in result2:
  print(elements[0], len(list(elements[1])))
```

### [5] 시간대별 가장 많이 팔린 품목 찾기

이번엔 시간대별로 가장 많이 팔린 품목을 찾고 시간대별로 정렬해 나타내봅시다. <br>
<b>Hint</b> max(X, key = X.count) : list X에 가장 많이 등장하는 element의 등장 횟수 반환

```python
result3 = bread.groupBy(lambda element:element[1][:2]).collect()
result3 = sorted(result3, key=lambda tup: tup[0])
for elements in result3:
  #print(elements[0], list(elements[1]))
  items = list(map(lambda x: x[3], list(elements[1])))
  print(elements[0], max(items, key=items.count))
  #print(elements[0], max(list(elements[1]), key=list(elements[1]).count)[3])
```

# NHL 게임 정보

### [1] game.csv 파일 읽기

```python
game = sc.textFile('/FileStore/tables/game.csv').map(lambda line : line.split(','))
```

### [2] 데이터프레임으로 바꾸어 형태 확인하기

```python
df_game = pd.DataFrame(game.collect())
df_game.head()
```

```python
# column 정보가 추가되어 있으므로 제거해봅시다
game = game.filter(lambda line: line[0] != '"game_id"')
```

### [3] 평균 총득점수가 많은 Top5 경기장 구하기

경기장에 따라 득점이 많이 나오기도 하고 적게 나오기도 합니다. <br>
점수가 많이 나는 경기장이 어디인지 찾아봅시다. <br>
단, <b> 경기 수가 10번보다 많은 팀</b>만을 대상으로 찾아봅시다. <br>
<br>
<b>Hint</b><br>
(총 득점수) = (home팀 득점수) + (away팀 득점수)

```python
result4 = game.groupBy(lambda element:element[10]).collect()
venues = []
for elements in result4:
  num_games = len(list(elements[1]))
  scores = list(map(lambda x: int(x[6])+int(x[7]) if num_games>10 else 0, list(elements[1])))
  average = sum(scores)/num_games
  venues.append((elements[0], average))
rdd_venue = sc.parallelize(venues)
rdd_venue.top(5, key=lambda x:x[1])
```

### [4] 홈경기 승률이 높은 Top5 팀 구하기

데이터에 각각의 게임 결과는 나와있지만 팀별 성적을 확인하기는 어렵네요. <br>
이번에는 팀별 홈경기 승률을 구해 상위 5개 팀을 알아봅시다. <br><br>

<b>Hint</b><br>
(홈경기 승률) = (이긴 홈경기) / (전체 홈경기)<br>
"outcome" 칼럼이 "home win -"이면 홈팀이 이긴 경기, "away win -"이면 홈팀이 진 경기입니다.
x.count('a'): list x에서 'a'의 등장횟수 반환

```python
result5 = game.groupBy(lambda element:element[5]).collect()
home_team_id = []
for elements in result5:
  num_games = len(list(elements[1]))
  num_win = list(map(lambda x: x[8][1], list(elements[1]))).count('h')
  winning_rate = num_win/num_games
  home_team_id.append((elements[0], winning_rate))
rdd_home = sc.parallelize(home_team_id)
rdd_home.top(5, key=lambda x:x[1])
  
```
