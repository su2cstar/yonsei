```python
%matplotlib inline

import warnings
from text_unidecode import unidecode
from collections import deque
warnings.filterwarnings('ignore')


import pandas as pd
from sklearn.manifold import TSNE
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from node2vec import Node2Vec

sns.set_style('whitegrid')
```

```python
data = pd.read_csv(r'C:\Users\ahn92\Documents\dev\Sources\20_Project\02_PJ_Embrain\cosTop3FromTo.csv')
```

```python
data.head()
```

```python
graph = nx.Graph()
```

```python
for i,j in zip(data['from'],data['to']):
    graph.add_edge(i,j)
```

```python
graph
```

```python
node2vec = Node2Vec(graph, dimensions=20, walk_length=16, num_walks=100, workers=2)
```

```python
model = node2vec.fit(window=10, min_count=1)
```

```python
model.wv.vocab
```

```python
for node, _ in model.most_similar('P0294908'):
    print(node)
```

```python
panel_nodes = [x for x in model.wv.vocab if len(x) > 3]
embeddings = np.array([model.wv[x] for x in panel_nodes])
```

```python
tsne = TSNE(n_components=2, random_state=7, perplexity=15)
embeddings_2d = tsne.fit_transform(embeddings)
```

```python
figure = plt.figure(figsize=(11, 9))

ax = figure.add_subplot(111)

ax.scatter(embeddings_2d[:, 0], embeddings_2d[:, 1])
```
