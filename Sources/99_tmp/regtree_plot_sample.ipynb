```python
import pandas as pd  
import numpy as np  
#import matplotlib.pyplot as plt  
#%matplotlib inline
#matplotlib 을 사용해서 plotting 하는 것이 아니기 때문에 matplotlib을 import 하지 않아도 동작합니다.
```

```python
dataset = pd.read_csv('harris.dat', header = None, delimiter=' ')  
```

```python
dataset.head()
```

```python
dataset.describe()  
```

```python
X = dataset.drop(0, axis=1) 
y = dataset[0]  
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
from sklearn.tree import DecisionTreeRegressor  
regressor = DecisionTreeRegressor(max_depth=3)  
regressor.fit(X_train, y_train)  
```

```python
y_pred = regressor.predict(X_test) 
```

```python
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})  
df  
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()
#export_graphviz 함수에서 생성되는 파일을 파이썬 메모리에 저장하기 위해 StringIO() 함수를 사용합니다

export_graphviz(regressor, out_file=dot_data, filled=True, rounded=True)
# .DOT 파일로 Decision Tree 모형을 저장합니다 위에서 사용한 dot_data 라는 변수에 저장을 합니다

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
# dot_data를 사용하여서 그래프를 생성합니다

Image(graph.create_png())
#Image 함수를 사용하여서 그래프를 표시합니다
```

```python
graph.write_png('tree.png')
# 파일로 저장하는것 또한 가능합니다
```

```python
## classifier에서도 사용 가능합니다.
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
iris = load_iris()
clf = clf.fit(iris.data, iris.target)
```

```python
dot_data2 = StringIO()
export_graphviz(clf, out_file=dot_data2, filled=True, rounded=True)
graph2 = pydotplus.graph_from_dot_data(dot_data2.getvalue())  
Image(graph2.create_png())
```
