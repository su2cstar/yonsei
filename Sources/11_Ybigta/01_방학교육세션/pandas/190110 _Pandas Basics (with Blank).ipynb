## numpy > list comprehension > for loop
---

for loop과 numpy의 연산 시간이 꽤 차이나는 것을 확인할 수 있습니다.<br>
pandas에서 Series에 for loop을 쓴다면 아마 영원히 없어지지 않는 * 를 보게 될 것입니다.<br>
웬만한 작업은 pandas의 method, 특히 groupby로 다 처리할 수 있습니다.<br>
<br>
<br>

```python
def sclar_vector_product(scalar, vector):
    result = []
    for value in vector:
        result.append(scalar * value)
    return result 

iternation_max = 100000

vector = list(range(iternation_max))
scalar = 2

%timeit sclar_vector_product(scalar, vector) # for loop을 이용한 성능
```

```python
%timeit [scalar * value for value in range(iternation_max)] # list comprehension을 이용한 성능
%timeit np.arange(iternation_max) * scalar # numpy를 이용한 성능
```

#### * read_csv 세부 설정
- header : 데이터에 column의 label이 있는 경우 default, 없을 경우 None으로 설정
- nrows : 불러올 행의 수. 지정하지 않을 경우 전체 데이터를 가져온다. (자주 쓰진 않아요. 데이터가 너무 클 때 적당히 끊는 용도로 씁니다.)

이처럼 확인된 결측치들은 어떻게 처리하면 좋을까요?<br>
데이터나 적용할 모델에 따라 다르겠으나, 그 수가 많지 않을 경우 그냥 0으로 채워주거나 결측치가 들어있는 row를 지워버립니다.<br>
특정 변수에만 결측치가 많다면 해당 column만을 지울 수도 있을 것입니다.<br>
또 다른 방법으로는 평균값,해당 데이터와 인접한 데이터의 값을 넣거나(KNN), 특정 모델로 예측한 값을 넣는 것도 생각해볼 수 있겠죠.<br>
일단 포켓몬 데이터의 결측치는 범주형 데이터이고 값이 없는게 정상이기 때문에 0으로 채우기, 행/열 지우기 정도만 연습해보겠습니다.


- <font color="#ff5b00">.dropna()</font>
- <font color="#ff5b00">.fillna()</font>

```python
# thresh -> 결측치가 n개 이상인 경우에만 drop

print(
    len(pokemon.dropna()),
    len(pokemon.dropna(thresh=2))) #살아남은 수를 보여줘요!
```

```python
# axis=1인 경우 열 기준으로 drop -> Type_2가 사라진 것을 확인 (하나라도 없는 열 drop)

pokemon.dropna(axis = 1).head()
```

```python
print(
    len(pokemon.dropna(axis = 1).columns),
    len(pokemon.columns))
```

```python
# 결측치를 0으로 채우기

pokemon.fillna(0).head()
```

```python
# 0이 아닌 다른 값으로도 채울 수 있음

pokemon.fillna("None").head()
```

# 5. Groupby
---
Groupby는 단순하게, **특정 기준으로 데이터를 분류하는 방법**으로 이해할 수 있습니다.<br>
예를 들어 그 분류기준이 색깔이라면, 색깔을 기준으로 그룹을 나누어서 색에 따른 각 포켓몬들의 특징을 확인할 수 있습니다.<br>

---
#### * 기본적인 사용 방법 : key를 기준으로 묶는다! -> 원하는 column 선택-> operation을 적용


같은 key를 가진 데이터끼리 묶어서
그룹마다 column을 뽑아 내서 (인덱싱) operation을 적용하여 계산한 뒤
각 그룹의 결과를 합쳐서 하나의 데이터프레임으로 만든다.

- <font color="#ff5b00">df.groupby(key)[column].operation()</font>
- 대표적인 operation의 종류
    - size() : 갯수
    - sum() : 합
    - mean() : 평균
    - median() : 중앙값
    - std() : 표준편차
    - max() : 최댓값
    - min() : 최솟값
    - apply() : 내가 만든 함수 적용
    - <font color="#ff5b00">agg(['function1', 'function2', ...])</font> : 여러 개의 함수를 동시에 적용!

pokemon을 [Generation]별로 묶어서 세대별 row의 [size]를 보고 싶다면?

```python
pokemon.groupby('Generation').size()
```

---
#### * grouped : groupby로 묶인 상태에서 데이터 다루기

- <font color="#ff5b00">df.get_group()</font>
- <font color="#ff5b00">filter()</font>
- <font color="#ff5b00">agg()의 활용</font>

```python
grouped = pokemon.groupby("Generation")
```

```python
# 총 6개의 그룹이 생성된 것을 확인

for name,group in grouped:
    print (type(name))
    print (type(group))
    print (len(group))
```

```python
# 각 그룹을 그 자체로 불러오는 방법

grouped.get_group(1).head()
```

```python
grouped.filter(lambda x: len(x) <= 120).head()
```

```python
# Total의 max 값이 600이상인 타입의 그룹들만 모아서 보여주기

pokemon.groupby("Type_1").filter(lambda x: x["Total"].max() > 600).head()
```

```python
# 이런 것까지 할 수 있다 정도만 알아두시면 좋습니다.

grouped = pokemon.groupby(['Type_1', 'isLegendary']).agg({'Height_m': "mean",     
                                     'Type_2': "nunique", 
                                     'Total': ['mean', 'max', 'var']})
grouped
```

```python
# 더블 인덱스가 보기싫어서 정리할 때 -> droplevel

grouped.columns = grouped.columns.droplevel(level=0)
grouped
```
