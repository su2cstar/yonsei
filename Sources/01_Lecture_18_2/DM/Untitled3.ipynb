```python
# -*- coding: utf-8 -*-
"""
Created on Thu Sep 20 13:50:02 2018

@author: 2018321084 안주영
"""
#%%
##그래플 출력을 위한 matplotlib import
import matplotlib.pyplot as plt
%matplotlib inline 
##데이터 input
titanic = [
[1,1,"Smith, Mrs. Lucien Philip","F",18],
[0,3,"Gustafsson, Mr. Karl Gideon","M",19],
[1,3,"Jalsevac, Mr. Ivan","M",29],
[0,2,"Hiltunen, Miss. Marta","F",18],
[0,2,"Fillbrook, Mr. Joseph Charles","M",18],
[1,3,"McCoy, Miss. Alicia","F",25],
[0,3,"Lindahl, Miss. Agda Thorilda Viktoria","F",25],
[0,2,"Butler, Mr. Reginald Fenton","M",25],
[0,3,"Harknett, Miss. Alice Phoebe","F",32],
[0,3,"O'Brien, Mr. Timothy","M",49],
[1,1,"Wilson, Miss. Helen Alice","F",31],
[0,3,"Sage, Mr. Frederick","M",38],
[0,3,"Palsson, Miss. Stina Viola","F",33],
[1,3,"O'Brien, Mrs. Thomas","F",37],
[1,3,"Persson, Mr. Ernst Ulrik","M",25],
[1,1,"Hays, Mrs. Charles Melville","F",52],
[0,3,"Vander Planke, Miss. Augusta Maria","F",18],
[0,3,"Brobeck, Mr. Karl Rudolf","M",22],
[1,1,"Salomon, Mr. Abraham L","M",63],
[0,1,"Chisholm, Mr. Roderick Robert Crispin","M",51],
[0,3,"Leinonen, Mr. Antti Gustaf","M",32],
]
# 리스트 데이터 구성 (왼쪽부터)
# 생존여부(1:생존, 0:비생존), 객실등급(1~3), 승객이름, 성별(‘M’:남, ‘F’:여), 나이

#%%
##1
#계산을 위한 함수 생성
# 해당 index 번호의 최대값을 retrun하는 함수
def getmax(lst,ind):
    mx = lst[0][ind]
    for i in range(len(lst)):
        if mx <= lst[i][ind]:
            mx = lst[i][ind]
    return mx
# 해당 index 번호의 최소값을 retrun하는 함수
def getmin(lst,ind):
    mn = lst[0][ind]
    for i in range(len(lst)):
        if mn >= lst[i][ind]:
            mn = lst[i][ind]
    return mn
# 해당 index 번호의 평균을 retrun하는 함수
def getmean(lst,ind):
    mn = 0
    for i in range(len(lst)):
        mn = mn + lst[i][ind]
    mn = mn / len(lst)
    return mn
# 해당 list를 해당 성별에 해당하는 값들만 갖는 list로 분할하는 함수
def spltsex(lst,sex='M',ind=3):
    spltlst=[]
    for i in range(len(lst)):
        if lst[i][ind]==sex:
            spltlst.append(lst[i])
    return spltlst

#주어진 리스트를 남 여 별로 분할
mlst, flst = spltsex(titanic), spltsex(titanic,'F')

# 생성한 함수를 이용하여 출력
print("(최연소) 남 : %d 여: %d" %(getmin(mlst,4),getmin(flst,4)))
print("(최고령) 남 : %d 여: %d" %(getmax(mlst,4),getmax(flst,4)))
print("(평균연령) 남 : %2.14f 여: %2.1f" %(getmean(mlst,4),getmean(flst,4)))

#%%
##2
# 람다함수를 이용하여 정렬을 하기위해 사용할 key 지정
srtlst = sorted(titanic, key = lambda x : x[4] , reverse = True)
# 3차원 리스트의 원소 하나씩 출력
for i in range(len(srtlst)):
    print(srtlst[i])

#%%    
##3
#unique한 객실 등급을 확인
grd=sorted(list(set([row[1] for row in titanic])))
grdlst = [row[1] for row in titanic]
#각 등급별 count 확인
cnt=[]
for i in grd:
    cnt.append(grdlst.count(i))
#그래프 출력을 위한 함수
def plot1():
    plt.bar(grd,cnt)
    plt.xticks(grd)
    plt.title('Titanic Passengers by Classes')
    plt.xlabel('Passenger Classes')
    plt.ylabel('# of Passengers')
    plt.show()
#그래프 출력
plot1()

#%%
##4
# list를 생존 비생존으로 분할해주는 함수
def spltsrv(lst,srv=1,ind=0):
    spltlst=[]
    for i in range(len(lst)):
        if lst[i][ind]==srv:
            spltlst.append(lst[i])
    return spltlst
# 나이를 10살 단위로 생존자/사망자별 list 생성
livelst, dielst = spltsrv(titanic) , spltsrv(titanic, srv = 0)
lvage , dieage = [row[4] for row in livelst] , [row[4] for row in dielst]
lvag10 , dieag10 = list(map(lambda x : (x//10)*10 , lvage)) , list(map(lambda x : (x//10)*10 , dieage))

#x축으로 사용하기 위해 10살 구간 나이를 최고령자나이대까지 list로 생성
ind = list(range((getmax(titanic,4)//10)+1))
ind10 = list(map(lambda x : x*10 , ind))

#각 구간별 count를 생존자 사망자별 list 생성
cntsrv , cntdie = [] , []
for i in ind10:
    cntsrv.append(lvag10.count(i))
    cntdie.append(dieag10.count(i))
#그래프 출력을 위한 함수 생성
def plotsrv():
    plt.bar(ind10,cntsrv,8)
    plt.xticks(ind10)
    plt.title('Titanic Passengers by Ages - Survived')
    plt.xlabel('Age Groups')
    plt.ylabel('# of Passengers')
    plt.show()

def plotdie():
    plt.bar(ind10,cntdie,8)
    plt.xticks(ind10)
    plt.title('Titanic Passengers by Ages - Not Survived')
    plt.xlabel('Age Groups')
    plt.ylabel('# of Passengers')
    plt.show()
#그래프 출력
plotsrv()
plotdie()

#%%
##5
#딕셔너리의 키로 사용할 값들을 list로 저장
dictkey = ['survived','pclass','name','sex','age']

#titanic list의 name 항을 ,를 구분자로 앞부분만 가져온다
name =[]
for i in range(len(titanic)):
    name.append(titanic[i][2].split(',')[0])
#앞에서 생성한 name list를 사용하여 list 생성
titadict={}
for i in range(len(titanic)):
    titadict[name[i]] = dict(zip(dictkey, titanic[i]))
#dictionary의 key 와 value를 한줄씩 출력
for i in titadict.keys():
    print(i,titadict[i])    

```
