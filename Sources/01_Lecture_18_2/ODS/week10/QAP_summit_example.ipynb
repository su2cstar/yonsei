```python
import pickle
articles_dict = pickle.load(open('Koreas_summit_dict.p', 'rb'))
for key in list(articles_dict)[:4]:
    print(articles_dict[key]['content'])
```

```python
kyunghyang_text = ''
hangyere_text = ''
donga_text =''
chosun_text = ''
for key in articles_dict:
    pub = articles_dict[key]['publisher']
    if pub == '경향신문':
        kyunghyang_text += '\n'+ articles_dict[key]['content']
    elif pub == '한겨레':
        hangyere_text += '\n'+ articles_dict[key]['content']
    elif pub == '동아일보':
        donga_text += '\n'+ articles_dict[key]['content']
    elif pub == '조선일보':
        chosun_text += '\n'+ articles_dict[key]['content']
```

```python
stopwords = ['하며','동아일보','경향신문','조선일보','단독', '핫한', 
             '한겨레', '재테크','재배포','동아닷컴','우리','사진','진행',
             '.co','.kr','나가기','페이스북','.com','오전','18일', '19일', 
             '이날','이번', '있다','홈페이지', '기자', '무단전재', '무단', 
             '보기경향신문', '클릭', '하기', '금지', '기자', '뉴스', '기사', 
             'AP통신', 'Copyrights', 'SNS', 'chosuncom', 'dongacom', 'hyopdhanicokr', 'tbs', 'wwwkhancokr']
```

```python
total_text = kyunghyang_text+' '+hangyere_text+' '+donga_text+' '+chosun_text
total_text = total_text.replace('\n', ' ')
```

명사를 추출해서 사전에 등록하기

```python
from kornounextractor.noun_extractor import extract
with open('dic.txt', 'w', encoding='utf8') as f:
    for word in sorted(extract(total_text, freq=1.0)):
        f.write(word+'\tNNG\n')
```

```python
import konlpy.tag
komoran = konlpy.tag.Komoran(userdic='dic.txt')
```

모든 신문기사에 사용된 명사 단어 추출하기

```python
Nouns = komoran.nouns(total_text)
final_nouns = Nouns.copy()
unique_nouns = set(Nouns)
for word in unique_nouns:
    # 1음절 단어와 불용어 제거하기
    if len(word) == 1:
        while word in final_nouns:
            final_nouns.remove(word)
    if word in stopwords:
        while word in final_nouns:
            final_nouns.remove(word)
```

```python
print(final_nouns)
```

```python
from kr_sna_QAP import do_kr_sna
NUM = 20
g_kh1 = do_kr_sna(kyunghyang_text, final_nouns, stopwords, num=NUM)
g_hgr1 = do_kr_sna(hangyere_text, final_nouns, stopwords, num=NUM)
g_da1 = do_kr_sna(donga_text,final_nouns, stopwords, num=NUM)
g_cs1 = do_kr_sna(chosun_text,final_nouns, stopwords, num=NUM)
```

전체 단어에서 상위 NUM 개의 단어들을 선택하고 해당 단어들(동일한 단어들)을 가지고 각 신문사의 신문기사를 이용해서 별도의 네트워크를 생성한다 <br>
네트워크 생성에 사용하는 단어의 선택 <br>
