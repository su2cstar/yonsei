```python
import requests
```

```python
url = 'https://www.daum.net/'
```

```python
r = requests.get(url) # ruquest 메세지를 보내는데 사용하는 함수 get 서버가 보낸 response message를 return 해 준다
```

```python
print(r.text) # 웹페이지를 구성하느 소스 코드 정보만 필요하다 text라는 세부 변수에 저장되어있다
```

```python
#beautiful soup을 통해서 소스에서 필요한 정보를 추출한다
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(r.text , 'lxml') # 여러개의 parser 중에서 lxml 으로 parsing 한다
```

```python
# tag의 이름을 사용하여 찾는다
soup('title')
soup.title
```

```python
#text정보만 추출하기위해서 text라는 키워드를 사용한다
soup.title.text
```
