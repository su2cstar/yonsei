```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import Select
```

```python
url = 'http://www.truedelta.com/'
dvpath = r'C:\Users\ahn92\Documents\Sources\ODS\chromedriver_win32\chromedriver.exe'
```

```python
driver = webdriver.Chrome(executable_path=dvpath)
driver.get(url)
driver.implicitly_wait(10)

select1 = Select(driver.find_element_by_name('results_page'))
select1.select_by_value('1') # value 값을 가지고 옵션 선택

select2 = Select(driver.find_element_by_name('brand'))
select2.select_by_value('Acura')

select3 = Select(driver.find_element_by_name('model'))
select3.select_by_value('293')

submit_element = driver.find_element_by_xpath('//*[@id="show-search"]/div[5]/input')

submit_element.click()

html = driver.page_source

driver.close()
```
