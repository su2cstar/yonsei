```python
import time
```

```python
def get_apt_info(data):
    total_apts = []
    data_dict = json.loads(data)
    for apt_dict in data_dict['articleInfo']['cfmArticleList']: # a list of dictionary, 20개 아파트의 정보
        sum_data = {}       
        desc = apt_dict['articleDescription']
        price = apt_dict['price']
        floor = apt_dict['floor']
        size1 = apt_dict['size1']
        size2 = apt_dict['size2']
        town_name = apt_dict['secName']
        sum_data['desc'] = desc
        sum_data['price'] = price
        sum_data['floor'] = floor
        sum_data['size1'] = size1
        sum_data['size2'] = size2
        sum_data['town_name'] = town_name
        total_apts.append(sum_data)
    return total_apts
```

```python
Referer= 'https://land.naver.com/search/article.nhn?tab=article&query=%EC%84%B8%EC%A2%85%ED%8A%B9%EB%B3%84%EC%9E%90%EC%B9%98%EC%8B%9C+%EC%95%84%ED%8C%8C%ED%8A%B8%20%EB%A7%A4%EB%A7%A4'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:51.0) Gecko/20100101 Firefox/51.0', 
    'Referer':Referer
}

total_data = []
for page in range(1,11):
    page = str(page)
    payload = {
    'page':page,
    'query':'세종특별자치시 아파트 매매',
    'regionCode':'3611000000',
    'regionType':'region2',
    'rletType':'A01:A03:A04',
    'tradeType':'A1'
    }
    
    url = 'https://land.naver.com/search/articleSearch.nhn' 
    r = requests.post(url, data = payload, headers = headers)
    total_data.extend(get_apt_info(r.text))
    time.sleep(1)
    
```
